# Titanic Analysis

## Introduction

##### The RMS Titanic was a British passenger liner that gained global fame due to its tragic sinking on its maiden voyage in April 1912. At the time, the Titanic was one of the largest and most luxurious ships ever built, often referred to as the "unsinkable" ship. The disaster occurred during the night of April 14-15, 1912, when the ship struck an iceberg in the North Atlantic Ocean.

##### The Titanic was part of the White Star Line's trio of Olympic-class ocean liners, which also included the RMS Olympic and HMHS Britannic. It was constructed in the Harland and Wolff shipyards in Belfast, Northern Ireland. The ship was heralded for its opulence and cutting-edge technology.

##### The Titanic set sail on its maiden voyage from Southampton, England, on April 10, 1912, with stops in Cherbourg, France, and Queenstown (now known as Cobh), Ireland, before heading across the Atlantic to its final destination, New York City. On the night of April 14, 1912, the Titanic struck an iceberg about 375 miles south of Newfoundland. The iceberg caused significant damage to the ship's hull, leading to its eventual sinking. The ship carried a limited number of lifeboats, and the evacuation process was chaotic, resulting in the loss of over 1500 lives.
##### The Titanic carried a diverse group of passengers, including wealthy industrialists, immigrants seeking a new life in America, and crew members. The class distinction among passengers became notable during the evacuation, as first-class passengers were given priority access to lifeboats.

##### The sinking of the Titanic captured global attention and shocked the world. The media extensively covered the disaster, and the public was deeply affected by the loss of life. The tragedy led to changes in maritime safety regulations and practices.
##### The sinking of the Titanic has become a symbol of human hubris and the unpredictable forces of nature. Numerous books, documentaries, and films have been produced about the Titanic, contributing to its enduring place in popular culture.

### Survival Analysis

#### a. What was the overall survival rate on the Titanic?

```python
overall_survival_rate = df['Survived'].mean()
print("Overall Survival Rate: {:.2%}".format(overall_survival_rate))
```

#### b. Did passengers in certain classes have a higher chance of survival?

```python
class_survival_rates = df.groupby('Pclass')['Survived'].mean()
print("Survival Rates by Passenger Class:")
print(class_survival_rates)
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rates by Passenger Class')
plt.show()
```

#### c. Did gender play a role in survival rates?

```python
gender_survival_rates = df.groupby('Sex')['Survived'].mean()
print("Survival Rates by Gender:")
print(gender_survival_rates)
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rates by Gender')
plt.show()
```

#### d. Were passengers with siblings or spouses more likely to survive?

```python
sibsp_survival_rates = df.groupby('SibSp')['Survived'].mean()
print("Survival Rates by SibSp (Siblings/Spouses Aboard):")
print(sibsp_survival_rates)
```

```python
sns.lineplot(x='SibSp', y='Survived', data=df)
plt.title('Survival Rates by SibSp (Siblings/Spouses Aboard)')
plt.show()
```

#### e. Did having parents or children aboard the Titanic influence survival rates?

```python
parch_survival_rates = df.groupby('Parch')['Survived'].mean()
print("Survival Rates by Parch (Parents/Children Aboard):")
print(parch_survival_rates)
```

```python
sns.barplot(x='Parch', y='Survived', data=df)
plt.title('Survival Rates by Parch (Parents/Children Aboard)')
plt.show()
```

### Age and Survival


#### a. What was the age distribution of passengers on the Titanic?

```python
plt.figure(figsize=(10, 6))
df['Age'].plot.hist(bins=20, edgecolor='k')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Passengers')
plt.show()
```

#### b. Did age impact a passenger's likelihood of survival?

```python
age_survival_rates = df.groupby('Age')['Survived'].mean()
print("Survival Rates by Age:")
print(age_survival_rates)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Survived', data=df)
plt.title('Survival Rates by Age')
plt.show()
```

#### c. Were children given priority in the lifeboats?

```python
child_survival_rate = df[df['Age'] < 18]['Survived'].mean()
print("Survival Rate for Children (Age < 18): {:.2%}".format(child_survival_rate))
```

#### Ticket Fare and Survival

```python
# a. What was the distribution of ticket fares?
df['Fare'].plot.hist(bins=20, edgecolor='k')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Ticket Fare Distribution')
plt.show()
```

#### b. Did passengers who paid more for their tickets have a higher chance of survival?

```python
fare_survival_rates = df.groupby('Fare')['Survived'].mean()
print("Survival Rates by Fare:")
print(fare_survival_rates)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fare', y='Survived', data=df)
plt.title('Survival Rates by Fare')
plt.show()
```

#### c. Did the ticket fare correlate with the passenger class?

```python
fare_class_correlation = df[['Fare', 'Pclass']].corr().iloc[0, 1]
print("Correlation between Fare and Pclass: {:.2f}".format(fare_class_correlation))
```

### Port of Embarkation

#### a. How many passengers embarked from each of the three ports (Southampton, Cherbourg, and Queenstown)?

```python
port_counts = df['Embarked'].value_counts()
print("Passengers Embarked from Each Port:")
print(port_counts)
```

#### b. Did the port of embarkation influence a passenger's survival chances?

```python
port_survival_rates = df.groupby('Embarked')['Survived'].mean()
print("Survival Rates by Port of Embarkation:")
print(port_survival_rates)
```

```python
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rates by Port of Embarkation')
plt.show()
```

### Family Size

#### a. What was the average family size on the Titanic?

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
average_family_size = df['FamilySize'].mean()
print("Average Family Size: {:.2f}".format(average_family_size))
```

#### b. Did passengers traveling alone have a different survival rate compared to those with family members aboard?

```python
alone_survival_rate = df[df['FamilySize'] == 0]['Survived'].mean()
with_family_survival_rate = df[df['FamilySize'] > 0]['Survived'].mean()
print("Survival Rate for Passengers Traveling Alone: {:.2%}".format(alone_survival_rate))
print("Survival Rate for Passengers with Family Aboard: {:.2%}".format(with_family_survival_rate))
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='FamilySize', y='Survived', data=df)
plt.title('Survival Rates by Family Size')
plt.show()
```

### Cabin and Deck Information

#### a. How many passengers had cabin information available in the dataset?

```python
cabin_info_available = df['Cabin'].count()
print("Passengers with Cabin Information Available: {}".format(cabin_info_available))
```

#### b. Did the deck level (A, B, C, etc.) where a passenger's cabin was located affect their survival?

```python
df['Deck'] = df['Cabin'].str.extract(r'([A-Z])')
deck_survival_rates = df.groupby('Deck')['Survived'].mean()
print("Survival Rates by Deck:")
print(deck_survival_rates)
```

```python
sns.barplot(x='Deck', y='Survived', data=df, order=sorted(df['Deck'].dropna().unique()))
plt.title('Survival Rates by Deck')
plt.show()
```

### Name Analysis

#### a. What are the most common titles (e.g., Mr., Mrs., Miss) among passengers, and did they impact survival?

```python
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.')
title_counts = df['Title'].value_counts()
print("Most Common Titles:")
print(title_counts)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Title', data=df, order=title_counts.index)
plt.title('Most Common Titles Among Passengers')
plt.xticks(rotation=45)
plt.show()
```

```python
title_survival_rates = df.groupby('Title')['Survived'].mean()
print("Survival Rates by Title:")
print(title_survival_rates)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Title', y='Survived', data=df, order=title_counts.index)
plt.title('Survival Rates by Title')
plt.xticks(rotation=45)
plt.show()
```
